```
# Import required libraries
import pandas as pd

# Define the correct file path
file_path = "/World-Stock-Prices-Dataset.csv"  # Replace with the actual path

# Step 1: Read the CSV file using pandas
df = pd.read_csv(file_path)
print("Dataset loaded successfully.")
print("Dataset loaded successfully.")
```

```
# Step 2: Display the first few rows of the DataFrame
print("First few rows of the dataset:")
print(df.head())
```

```
# Step 3: Check for missing values in the dataset
print("\nMissing values in the dataset:")
print(df.isnull().sum())
```

```
# Step 4: Check data types of all columns
print("\nData types of the columns:")
print(df.dtypes)
```

```
# Step 5: Display the first 10 rows of the dataset
print("\nFirst 10 rows of the dataset:")
print(df.iloc[:10])
```

```
# Step 6: Handle non-numeric columns (convert to numeric where necessary)
df['Volume'] = pd.to_numeric(df['Volume'], errors='coerce')  # Coerce errors to NaN
df['Capital Gains'] = pd.to_numeric(df['Capital Gains'], errors='coerce')

# Step 7: Re-check missing values after conversion
print("\nMissing values after conversion:")
print(df.isnull().sum())
```

```
# Step 8: Display summary statistics for numeric columns
print("\nSummary statistics for numeric columns:")
print(df.describe())
# Step 9: Correlation Analysis
# Select numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate correlation matrix
correlation_matrix = df[numeric_cols].corr()

# Display correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)

# Visualize correlation with a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```
# Step 9: Group by 'Industry_Tag' and calculate the total 'Volume' for each industry
industry_volume = df.groupby('Industry_Tag')['Volume'].sum().sort_values(ascending=False)
print("\nTotal Volume by Industry:")
print(industry_volume)
```

```
# Step 10: Plot the total volume by industry (logarithmic scale)
plt.figure(figsize=(10, 6))
sns.barplot(x=industry_volume.index, y=industry_volume.values)
plt.title('Total Volume by Industry')
plt.xlabel('Industry')
plt.ylabel('Total Volume')
plt.xticks(rotation=90)
plt.yscale('log')  # Logarithmic scale for better visualization of large differences
plt.show()
```

```
# Step 11: Create a pie chart for volume distribution by industry
plt.figure(figsize=(8, 8))
plt.pie(industry_volume.values, labels=industry_volume.index, autopct='%1.1f%%', startangle=140)
plt.title('Volume Distribution by Industry')
plt.show()
```

```
# Step 12: Get top 10 industries by total volume
top_industries = industry_volume.head(10)
# Step 13: Plot the top 10 industries by total volume
plt.figure(figsize=(10, 6))
sns.barplot(x=top_industries.index, y=top_industries.values)
plt.title('Top 10 Industries by Total Volume')
plt.xlabel('Industry')
plt.ylabel('Total Volume')
plt.xticks(rotation=45)
plt.yscale('log')  # Optional: Use logarithmic scale if values vary greatly
plt.show()
```

```
# Additional: Plot volume column as a bar graph (if meaningful)
if 'Volume' in df.columns:
    plt.figure(figsize=(12, 6))
    df['Volume'].head(20).plot(kind='bar', title='Stock Volume for First 20 Entries')
    plt.ylabel('Volume')
    plt.show()
```

```
# Top 5 industries by total stock volume
top_5_industries = df.groupby('Industry_Tag')['Volume'].sum().nlargest(5) # Changed 'industry' to 'Industry_Tag'
print("Top 5 Industries by Total Stock Volume:")
print(top_5_industries)

# Visualization - Bar Chart
top_5_industries.plot(kind='bar', color='skyblue', figsize=(10, 6), title='Top 5 Industries by Volume')
plt.xlabel('Industry')
plt.ylabel('Total Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```
# Find the industry with the highest stock volume
highest_volume_industry = df.groupby('Industry_Tag')['Volume'].sum().idxmax()
highest_volume_value = df.groupby('Industry_Tag')['Volume'].sum().max()

print("Industry with the Highest Stock Volume:")
print(f"Industry: {highest_volume_industry}")
print(f"Volume: {highest_volume_value}")

# Visualization - Bar Chart
df.groupby('Industry_Tag')['Volume'].sum().plot(kind='bar', color='lightgreen', figsize=(10, 6), title='Stock Volume by Industry')
plt.xlabel('Industry')
plt.ylabel('Total Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
# Average stock volume for all industries
average_volume = df['Volume'].mean()
print(f"Average Stock Volume for All Industries: {average_volume}")

# Visualization - Bar Chart for Average Volume
plt.figure(figsize=(6, 4))
plt.bar('Average Volume', average_volume, color='lightcoral')
plt.title('Average Stock Volume for All Industries')
plt.ylabel('Average Volume')
plt.tight_layout()
plt.show()

```

```
# Industry with the lowest stock volume
lowest_volume_industry = df.groupby('Industry_Tag')['Volume'].sum().idxmin()
lowest_volume_value = df.groupby('Industry_Tag')['Volume'].sum().min()

print(f"The industry with the lowest stock volume is: {lowest_volume_industry} with a volume of {lowest_volume_value}")

# Visualization - Line Chart for the lowest volume industry
plt.figure(figsize=(6, 4))
plt.plot([lowest_volume_industry], [lowest_volume_value], marker='o', color='lightcoral', linestyle='-', linewidth=2, markersize=8)

# Adding details to the line chart for better clarity
plt.title('Industry with the Lowest Stock Volume')
plt.xlabel('Industry')
plt.ylabel('Stock Volume')
plt.xticks([lowest_volume_industry])  # Show the industry on the x-axis
plt.yticks([lowest_volume_value])  # Show the stock volume on the y-axis
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()

```

```
import matplotlib.pyplot as plt

# Plot histogram for stock volume distribution across industries
plt.figure(figsize=(10, 6))
df['Volume'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black', alpha=0.7)

plt.title('Stock Volume Distribution Across Industries', fontsize=16)
plt.xlabel('Stock Volume', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()

plt.show()
```

```
# Get the top 10 industries by total stock volume
top_10_industries = df.groupby('Industry_Tag')['Volume'].sum().nlargest(10)

# Plot boxplot for the volume distribution of the top 10 industries
plt.figure(figsize=(10, 6))
top_10_industries.plot(kind='box', vert=False, color='skyblue')

# Add labels and title
plt.title('Stock Volume Distribution for Top 10 Industries', fontsize=16)
plt.xlabel('Stock Volume', fontsize=12)
plt.tight_layout()

plt.show()
```

```
# Get top 5 industries with the highest stock volume
top_5_high_volumes = df.groupby('Industry_Tag')['Volume'].sum().nlargest(5)

# Get top 5 industries with the lowest stock volume
top_5_low_volumes = df.groupby('Industry_Tag')['Volume'].sum().nsmallest(5)

# Create a figure for the bar chart
plt.figure(figsize=(10, 6))

# Plot the industries with the highest stock volumes
top_5_high_volumes.plot(kind='bar', color='green', alpha=0.7, label='High Volume', position=1)

# Plot the industries with the lowest stock volumes
top_5_low_volumes.plot(kind='bar', color='red', alpha=0.7, label='Low Volume', position=0)

# Add labels and title
plt.title('Stock Volume Comparison for Industries with Extremely High and Low Volumes', fontsize=16)
plt.xlabel('Industry', fontsize=12)
plt.ylabel('Stock Volume', fontsize=12)
plt.xticks(rotation=45)
plt.legend()

# Adjust layout to make it look better
plt.tight_layout()

# Show the plot
plt.show()
```

```
# Calculate the median volume for each industry
median_volume_per_industry = df.groupby('Industry_Tag')['Volume'].median()

# Display the median volumes
print("Median Volume per Industry:")
print(median_volume_per_industry)

# Visualization - Bar Chart for Median Volumes
plt.figure(figsize=(12, 6))
median_volume_per_industry.sort_values(ascending=False).plot(kind='bar', color='purple', alpha=0.8)
plt.title('Median Stock Volume per Industry', fontsize=16)
plt.xlabel('Industry', fontsize=12)
plt.ylabel('Median Stock Volume', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
# Total stock volume across all industries
total_volume = df['Volume'].sum()

# Calculate cumulative percentage contribution by industry
industry_volume = df.groupby('Industry_Tag')['Volume'].sum().sort_values(ascending=False)
industry_cumulative_percent = (industry_volume.cumsum() / total_volume) * 100

# Filter industries contributing to over 50% of total volume
industries_over_50_percent = industry_cumulative_percent[industry_cumulative_percent > 50]
print("Industries Contributing to Over 50% of Total Volume:")
print(industries_over_50_percent)

# Visualization - Bar Chart
plt.figure(figsize=(10, 6))
industries_over_50_percent.plot(kind='bar', color='orange')
plt.title("Industries Contributing Over 50% of Total Volume")
plt.xlabel("Industry")
plt.ylabel("Cumulative Percentage (%)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```
# Volume for 'technology' and 'automotive' industries
selected_industries = ['technology', 'automotive']
volume_selected_industries = df[df['Industry_Tag'].isin(selected_industries)].groupby('Industry_Tag')['Volume'].sum()

print("Volume for 'technology' and 'automotive' industries:")
print(volume_selected_industries)

# Visualization - Bar Chart
plt.figure(figsize=(8, 5))
volume_selected_industries.plot(kind='bar', color=['blue', 'green'])
plt.title("Volume for 'Technology' and 'Automotive' Industries")
plt.xlabel("Industry")
plt.ylabel("Total Volume")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```
# Identify the top 3 industries by total stock volume
top_3_industries = df.groupby('Industry_Tag')['Volume'].sum().nlargest(3).index

# Filter data for only the top 3 industries
top_3_data = df[df['Industry_Tag'].isin(top_3_industries)]

# Group by 'Date' and 'Industry_Tag' to calculate total volume per industry over time
volume_trend = top_3_data.groupby(['Date', 'Industry_Tag'])['Volume'].sum().reset_index()

# Pivot the data for better visualization
pivot_trend = volume_trend.pivot(index='Date', columns='Industry_Tag', values='Volume')

# Plot the trends
pivot_trend.plot(figsize=(12, 6), title="Total Stock Volume Trend for Top 3 Industries")
plt.xlabel("Date")
plt.ylabel("Total Volume")
plt.grid(alpha=0.5)
plt.legend(title="Industry", loc='upper left')
plt.tight_layout()
plt.show()
```
